**EJERCICIO 2**

```python
from math import sqrt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#Creamos un arreglo de 0 a 150
A=[]
for i in range(150):
  A.append(np.random.randint(0,150))


#Para calcular el promedio definimos una función
def promedio(A):
#Definimos la suma de cada uno de los elementos del arreglo y lo dividimos en el número de elementos que tiene el arreglo
  suma=0
  for i in A:
    suma=i+suma
    n=len(A)
    promedio_1=suma/n
  print(promedio_1)
promedio(A)
#Para calcular la varianza definimos una función
def varianza(A):
#Como la formula de la varianza es la sumatoria de los (xi-promedio) al cuadrado sobre el número de elementos del arreglo entonces definimos cada uno de ellos
  suma=0
  suma_varianza=0
  varianza=0
  n=len(A)
  for i in A:
#Primero se calcula el promedio como se mostró anteriormente
    suma=i+suma
    media=suma/n
#Luego se define la suma de cada elemento menos el promedio y todo eso al cuadrado
    suma_varianza=(i-media)**2+suma_varianza
#Luego dividimos la suma anterior sobre el número de elementos del arreglo, lo cual resulta la varianza por la fórmula anterior
    varianza=suma_varianza/n
  print(varianza)
varianza(A)
#Para calcular la desviación estándar primero definimos una función
def desviacion_estandar(A):
#Notemos que la desviación estandar es la raiz cuadrada de la varianza, por lo tanto calculamos la varianza y le sacamos raíz cuadrada
  suma=0
  suma_varianza=0
  varianza=0
  n=len(A)
  for i in A:
    suma=i+suma
    media=suma/n
    suma_varianza=(i-media)**2+suma_varianza
    varianza=suma_varianza/n
  print(sqrt(varianza))
desviacion_estandar(A)
```

```python
#Después graficamos el histograma
plt.hist(A,color="pink")
plt.grid(color = "yellow", linestyle = "--", linewidth = 1.5)
plt.title("Histograma del arreglo A")
plt.ylabel("Frecuencias")
plt.xlabel("Intervalos de números enteros de 0 al 150")
plt.show()
```

```python
#La suma de los pares
suma_pares=0
suma_impares=0
for i in A:
  if i%2==0:
    suma_pares=i+suma_pares
print("La suma de los pares es ",suma_pares)
#La suma de los impares
for i in A:
  if i%2!=0:
    suma_impares=i+suma_impares
print("La suma de los impares es ",suma_impares)
```

**EJERCICIO 3**

```python
import numpy as np
import matplotlib.pyplot as plt
#Primero definimos la funcion solicitada de la manera que sigue
def funcion(x,C):
  return (x-1) + C*np.exp(-x)
#Definimos valores para x
x = np.linspace(-10, 10, 50)
#Ahora veamos los valores de C>0
C_mayor = range(1,11)
valores_y1=[funcion(x, C) for C in C_mayor]
#Ahora veamos los valores de C<0
C_menor = range(-11,-1)
valores_y2=[funcion(x, C) for C in C_menor]
#Ahora los valores de C=0
C=[]
for i in range(50):
  C.append(0)
#Ahora graficamos
for i in range(len(C_mayor)):
    plt.plot(x, valores_y1[i],color="pink",label="C>0")
for i in range(len(C_menor)):
    plt.plot(x, valores_y2[i],color="yellow",label="C<0")
plt.plot(x, funcion(x,C),"--k",label="C=0")
plt.grid(color = "purple", linestyle = ":", linewidth = 1)
plt.title("Gráfica de la función")
plt.ylabel("y")
plt.xlabel("x")
plt.legend()
plt.show()
```
