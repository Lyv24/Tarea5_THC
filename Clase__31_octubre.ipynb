```python
def grafica(x):
  x = np.linspace(0, 6*np.pi ,100)
  seno = np.sin(x)

  plt.subplot(1,2,1)
  plt.plot(x,seno, "m", ms = 5)
  plt.grid(color = "green", linestyle = "--", linewidth = 1.5)
  plt.title("Gráfica de seno")

  x = np.linspace(0, 6*np.pi ,100)
  coseno = np.cos(x)

  plt.subplot(1,2,2)
  plt.plot(x,coseno, "--k", ms = 5)
  plt.grid(color = "pink", linestyle = "--", linewidth = 1.5)
  plt.title("Gráfica de coseno")
  plt.suptitle("Funciones trigonométricas")
  plt.show()

grafica(5)
```
