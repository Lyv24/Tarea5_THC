```python

x = np.linspace(0, 6*np.pi ,100)
seno = np.sin(x)
coseno = np.cos(x)

plt.plot(x,seno,":m", ms = 5)
plt.plot(x,coseno,":", ms = 5)
plt.title("Gr√°fica de coseno y seno")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid(color = "pink", linestyle = "--", linewidth = 1.5)#Pone la cuadr√≠cyla, cambia el color, el tipo de l√≠nea y el grosor
plt.show()
```
